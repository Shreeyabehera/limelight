```python
import pandas as pd
import re                   #likely won't use in scrape, just analysis
import requests
from bs4 import BeautifulSoup as BS
from IPython.core.display import HTML      #likely won't use
from datetime import datetime, timedelta
```

# Most recent data

50 weeks with last week ending 2023-05-14

```python
week_ending_str = '2023-05-14'
week_ending_field = datetime.strptime(week_ending_str, '%Y-%m-%d')
modifier = timedelta(7)
```

```python
title = []
weekly_gross = []        #formerly weekblah
gross_difference = []    #may not work, will likely need calculated column
average_tix_price = []
seats_sold = []
performance_count = []
percent_capacity = []
capacity_difference = []
week_ending = []
```

```python
i = 1
while i in range (1, 53):       #get data for last 52 weeks
    print(i)
    print(week_ending_field)
    url = f'https://www.playbill.com/grosses?week={str(week_ending_field.date())}'
    print(url)
    
    data = requests.get(url).text
    soup = BS(data, 'html.parser')
    table = soup.find('table', attrs={'class':'bsp-table'})
    
    for row in table.tbody.find_all('tr'):
        columns = row.find_all('td')
        if(columns != []):
            
            show = columns[0].text.strip()
            title.append(show)
            
            gross = columns[1].text.strip()
            weekly_gross.append(gross)
            
            grossdif = columns[2].text.strip()
            gross_difference.append(grossdif)
            
            avgtix = columns[3].text.strip()
            average_tix_price.append(avgtix)
            
            seats = columns[4].text.strip()
            seats_sold.append(seats)
            
            perfs = columns[5].text.strip()
            performance_count.append(perfs)
            
            perccap = columns[6].text.strip()
            percent_capacity.append(perccap)
            
            perccapdif = columns[7].text.strip()
            capacity_difference.append(perccapdif)
            
            week_ending.append(week_ending_field)
    i += 1
    week_ending_field = week_ending_field - modifier
```

```python
df = pd.DataFrame({'Title' : title, 'Weekly Gross' : weekly_gross, 'Gross Difference' : gross_difference, 'Average Ticket Price' : average_tix_price, 'Seats Sold' : seats_sold, 'Performance Count' : performance_count, 'Percent Capacity' : percent_capacity, 'Capacity Difference' : capacity_difference, 'Week Ending' : week_ending})
df[['Title', 'Theater']] = df['Title'].str.split('\n\n', expand=True)
#df[['Weekly Gross', 'Potential Gross']] = df['Weekly Gross'].str.split('\n', expand=True) #this one might not be necessary or do anything
df[['Average Ticket Price', 'Top Ticket Price']] = df['Average Ticket Price'].str.split('\n', expand=True)
df[['Seats Sold', 'Theater Capacity']] = df['Seats Sold'].str.split('\n', expand=True)
df[['Performance Count', 'Preview Count']] = df['Performance Count'].str.split('\n', expand=True)
```

```python
df.to_csv('../data/playbill-2023-05-14-to-2022-05-22.csv')
```

# Pre-Pandemic

```python
pre_week_ending_str = '2020-03-08'
pre_week_ending_field = datetime.strptime(pre_week_ending_str, '%Y-%m-%d')
modifier = timedelta(7)
```

```python
pre_title = []
pre_weekly_gross = []        #formerly weekblah
pre_gross_difference = []    #may not work, will likely need calculated column
pre_average_tix_price = []
pre_seats_sold = []
pre_performance_count = []
pre_percent_capacity = []
pre_capacity_difference = []
pre_week_ending = []
```

```python
# Warning: running this cell will take a while (it took ~35 minutes)
i = 1
while i in range (1, 532):       #need 531 weeks pre-pandemic shutdown
#while i in range (1, 45):       #need 45 weeks pre-pandemic shutdown     
    print(i)
    print(pre_week_ending_field)
    url = f'https://www.playbill.com/grosses?week={str(pre_week_ending_field.date())}'
    print(url)
    
    pre_data = requests.get(url).text
    pre_soup = BS(pre_data, 'html.parser')
    pre_table = pre_soup.find('table', attrs={'class':'bsp-table'})
    
    for row in pre_table.tbody.find_all('tr'):
        columns = row.find_all('td')
        if(columns != []):
            
            pre_show = columns[0].text.strip()
            pre_title.append(pre_show)
            
            pre_gross = columns[1].text.strip()
            pre_weekly_gross.append(pre_gross)
            
            pre_grossdif = columns[2].text.strip()
            pre_gross_difference.append(pre_grossdif)
            
            pre_avgtix = columns[3].text.strip()
            pre_average_tix_price.append(pre_avgtix)
            
            pre_seats = columns[4].text.strip()
            pre_seats_sold.append(pre_seats)
            
            pre_perfs = columns[5].text.strip()
            pre_performance_count.append(pre_perfs)
            
            pre_perccap = columns[6].text.strip()
            pre_percent_capacity.append(pre_perccap)
            
            pre_perccapdif = columns[7].text.strip()
            pre_capacity_difference.append(pre_perccapdif)
            
            pre_week_ending.append(pre_week_ending_field)
    i += 1
    pre_week_ending_field = pre_week_ending_field - modifier
```

```python
print(len(pre_title))
print(len(pre_weekly_gross))        #formerly weekblah
print(len(pre_gross_difference))    #may not work, will likely need calculated column
print(len(pre_average_tix_price))
print(len(pre_seats_sold))
print(len(pre_performance_count))
print(len(pre_percent_capacity))
print(len(pre_capacity_difference))
print(len(pre_week_ending))
```
